```python
###📊 Data Visualization Graph Types – Notes
###1. Line Plot
Use case: Continuous data over time or sequential data.
Syntax: plt.plot(x, y)
Ideal for: Trend analysis, time series, sensor data.
    

###2. Bar Graph
Use case: Comparing categorical data.
Syntax: plt.bar(categories, values)
Ideal for: Sales comparisons, counts per group, frequency.
    

##3. Heatmap
Use case: Representing matrix values with color intensity.
Syntax: plt.imshow(data, cmap='hot') or with seaborn: sns.heatmap(data)
Ideal for: Correlation matrices, confusion matrices, intensity maps.
    

###4. Geographical Plot
Use case: Mapping data to geographical locations.
Tools: Basemap, Cartopy, or Plotly for choropleth, scatter_geo
Ideal for: Population density, sales by region, climate data.
    

###5. Box Plot
Use case: Distribution, spread, outliers of numeric data.
Syntax: plt.boxplot(data) or sns.boxplot(x)
Ideal for: Statistical summaries, comparing distributions.
    

###6. Regression Plot (y = mx + c)
Use case: Show trend line over scatter plot.
Syntax: sns.regplot(x=x, y=y) (Seaborn)
Ideal for: Linear relationship modeling.
    

###7. Violin Plot
Use case: Distribution + KDE + boxplot in one view.
Syntax: sns.violinplot(x=data)
Ideal for: Comparing multiple distributions with more detail than boxplot.
    
###8. Swarm Plot
Use case: Scatterplot with no overlapping, great for small datasets.
Syntax: sns.swarmplot(x=data)
Ideal for: Showing all data points in a category.
    
###9. Rug Plot
Use case: Small vertical lines for each data point along an axis.
Syntax: sns.rugplot(data)
Ideal for: Distribution overview, supplementing KDE/Hist.
    
###10. Pie Chart
Use case: Proportion of parts to a whole.
Syntax: plt.pie(values, labels=labels)
Ideal for: Simple categorical data representation.
    

###11. Donut Chart
Use case: Like pie chart, with a central hole.
Syntax: plt.pie(...); plt.gca().add_artist(plt.Circle(...))
Ideal for: Enhanced aesthetic of pie chart.
    

###12. Sunburst Chart
Use case: Hierarchical data shown as nested sectors.
Tool: Plotly → plotly.express.sunburst()
Ideal for: Multi-level category representation.
    
###13. Surface Plot
Use case: 3D representation of z = f(x, y)
Tool: Matplotlib → ax.plot_surface(X, Y, Z)
Ideal for: Elevation, topography, complex 3D functions.
```
